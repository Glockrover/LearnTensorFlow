What I have learned so far:
1. Tensors are the main building blocks of TensorFlow.
2. A tensor is a generalization of vectors and matrices to potentially higher dimensions.
3. The number of dimensions of a tensor is called its rank.
4. The shape of a tensor is the number of elements in each dimension.
5. The data type of a tensor is the type of data it contains (e.g., float32, int32, etc.)

* scalar: a single number
* vector: a number with direction (e.g., wind speed and direction)  
* matrix: a 2-dimensional array of numbers
* Tensor: an n-dimensional array of numbers (where n can be any number, a 0-dimensional tensor is a scalar and 1-dimenstional tensor is a vector)

### Creating tensors with `tf.Variable`

For more info on tf.Variable visit: https://www.tensorflow.org/api_docs/python/tf/Variable

```python
# Creating tensors with tf.Variable() as above we have created tensors with tf.constant() which are immutable (cannot be changed) but if we want to create a tensor which can be changed then we can use tf.Variable().

changeable_tensor = tf.Variable([10, 7])
unchangeable_tensor = tf.constant([10, 7])

changeable_tensor, unchangeable_tensor
```

```python
# Trying to change the first element of our changeable tensor
# changeable_tensor[0] = 7

# We get a ResourceVariable object which is not subscriptable (cannot be indexed) to change the value of a tensor we need to use the assign() method
changeable_tensor[0].assign(7)
changeable_tensor
```

### Trying to change the unchangeable tensor

`unchangeable_tensor[0] = 7` will give an error because we cannot change the value of a tensor created with tf.constant() as they are immutable (cannot be changed) but tensors created with tf.Variable() are mutable (can be changed).
`unchangeable_tensor.assign(7)` 

#### We get an error because we cannot change the value of a tensor created with tf.constant() as they are immutable (cannot be changed) but tensors created with tf.Variable() are mutable (can be changed).

#### Why would we want to change our tensors? When we writing/coding neural networks we might want some tensors to be change and others to be unchange but the catch is that the decision will be made for us behind the scene. 

🔑 **Note:** Rarely in practice will you need to decide wheter to use `tf.constant` or `tf.Variable` to create tensors, as TensorFlow does this for you.

### Creating random tensors
Random tensors are tensors of some arbitrary size which contain random numbers.
Learn more about random tensors here: https://www.tensorflow.org/api_docs/python/tf/random/uniform


```python
# Create a random tensor with values between 0 and 1 of size (3, 4)
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3, 4))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42).normal(shape=(3, 4))
random_2
```

```python
# To check if they are the same
random_1 == random_2
```

#### Why is it important to use a random tensor with a seed?
Using a seed ensures reproducibility of results. Without a seed, each time you run the code, you'll get different random numbers, making it difficult to reproduce results or debug issues.

### Getting information from tensors

When dealing with tensors you must be aware of the following attributs: 

* Shape: The length (number of elements) of each dimension of a tensor is called its shape. You can get the shape of a tensor with the `tf.shape` attribute.
* Rank: The number of dimensions a tensor has is called its rank. You can get the rank of a tensor with the `tf.rank()` function.
* Axis or Dimension: A particular dimension of a tensor is called an axis. You can get the number of axes (or dimensions) of a tensor with the `tf.ndim` attribute.
* Size: The total number of items in the tensor is called its size. You can get the size of a tensor with the `tf.size()` function.

```python
# Creating a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=(2, 3, 4, 5))
rank_4_tensor
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Size of tensor:", tf.size(rank_4_tensor))
print("Elements in the 0 axis:", rank_4_tensor.shape[0])
print("Elements in the last axis:", rank_4_tensor.shape[-1])
print("The total number of elements in the tensor:", tf.size(rank_4_tensor).numpy())
```

### Indexing tensors

Tensors can be indexed just like Python lists.

```python
# Get the first 2 elements of each dimension of our rank 4 tensor
rank_4_tensor[:2, :2, :2, :2]
```

```python
# Get the first element from each dimension from each index except for the last one
rank_4_tensor[:1, :1, :1, :]
```

```python
# Creating a rank 2 tensor (2 dimensions)
rank_2_tensor = tf.zeros(shape=(3, 4))
tf.rank(rank_2_tensor)
rank_2_tensor
```

```python
# Get the last item of each of row of our rank 2 tensor
rank_2_tensor[:, -1]
```

```python
# Adding in extra dimensions to our rank 2 tensor
rank_3_tensor = rank_2_tensor[..., tf.newaxis]
rank_3_tensor
```

```python
# Alternative to tf.newaxis
tf.expand_dims(rank_2_tensor, axis=-1) # add a new axis at the end (same as tf.newaxis)
tf.expand_dims(rank_2_tensor, axis=0) # add a new axis at the beginning
```

### Manipulating tensors (tensor operations)

**Basic operations**

`+`, `-`, `*`, `/`

```python
# You can add values to a tensor using the addition operator
tensor = tf.constant([[1, 2],
                      [3, 4]])
tensor + 10
```

```python
# Subtraction
tensor - 10
```

```python
# We can use the tensorflow built-in functions to perform operations on our tensor
# if you want to have the advantage of using TensorFlow's built-in functions, 
# you can use them instead of the operators. They often have additional features and can be more efficient.
print(tf.multiply(tensor, 10)) # same as tensor * 10
print(tf.add(tensor, 10)) # same as tensor + 10
print(tf.subtract(tensor, 10)) # same as tensor - 10
print(tf.divide(tensor, 10))  # same as tensor / 10 
```

### **Matrix multiplication**
In machine learning, matrix multiplication is one of the most common operations. It is used in many algorithms, including neural networks. In TensorFlow, you can perform matrix multiplication using the `tf.matmul()` function or the `@` operator.

For more info visit: https://www.tensorflow.org/api_docs/python/tf/linalg/matmul

There are two rules our tensors (or matrices) need to fulfill if we're going to matrix multiply them: 

* The inner dimensions must match.
* The resulting matrix has the shape of the outter dimensions.

```python
# Matrix multiplication in TensorFlow
print(tensor)
tf.matmul(tensor, tensor) # matrix multiplication between tensor and itself
```

```python
# Matrix multiplication with Python's operator @
tensor @ tensor
```

```python
A = tf.constant([[1, 2, 5], [7, 2, 1], [3, 3, 3]])
B = tf.constant([[3, 5], [6, 7], [1, 8]])

tf.matmul(A, B), A @ B
```

**The dot product**

Matrix multiplication is also referred to as the dot product. 
The dot product is a fundamental operation in linear algebra and is used in various applications, including machine learning and deep learning. It involves multiplying corresponding elements of two matrices and summing the results to produce a new matrix.

You can perform matrix multiplication using:

- The `@` operator (e.g., `X @ Y`)
- `tf.matmul()` function (e.g., `tf.matmul(X, Y)`)
- `tf.tensordot()` function (e.g., `tf.tensordot(X, Y, axes=1)`)

For more info: https://www.tensorflow.org/api_docs/python/tf/tensordot

```python
# Perform the dot product on X and Y (requires X or Y to to be transposed )
tf.tensordot(X, tf.transpose(Y), axes=1)
```

```python
# Perform matrix multiplication between X and Y (requires X or Y to be transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# Perform matrix multiplication between X and Y (requires X or Y to be reshaped)
tf.matmul(X, tf.reshape(Y, (2, 3)))
```

```python
# Check the values of Y, reshape Y and transpose Y
print("Normal Y:")
print(Y, "\n")
print("Reshaped Y:")
print(tf.reshape(Y, (2, 3)))
print()
print("Transposed Y:")
print(tf.transpose(Y))
```
