```python
import tensorflow as tf

print(tf.__version__)
```

```python
# Create a vector, scalar, matrix and tensor with values of your choosing using tf.constant() 

scalar = tf.constant(42) # Scalar has 0 dimensions
vector = tf.constant([1, 2, 3, 4, 5]) # Vector has 1 dimension
matrix = tf.constant([[1, 2], [3, 4]]) # Matrix has 2 dimensions 
tensor = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]) # Tensor has 3 dimensions

print("Scalar:", scalar, "\n")
print("Vector:", vector, "\n")
print("Matrix:", matrix, "\n")
print("Tensor:", tensor, "\n")

scalar.ndim, vector.ndim, matrix.ndim, tensor.ndim
```

```python
# Find the shape, rank and size of the tensors you created in 1
print("Scalar shape:", scalar.shape, "Rank:", scalar.ndim, "Size:", tf.size(scalar).numpy(), "\n")
print("Vector shape:", vector.shape, "Rank:", vector.ndim, "Size:", tf.size(vector).numpy(), "\n")
print("Matrix shape:", matrix.shape, "Rank:", matrix.ndim, "Size:", tf.size(matrix).numpy(), "\n")
print("Tensor shape:", tensor.shape, "Rank:", tensor.ndim, "Size:", tf.size(tensor).numpy(), "\n")
```

```python
# Create two tensors containing random values between 0 and 1 with shape [5, 300]
random_tensor1 = tf.random.uniform(shape=[5, 300], minval=0, maxval=1)
random_tensor2 = tf.random.uniform(shape=[5, 300], minval=0, maxval=1)
random_tensor1, random_tensor2
```

```python
# Multiply the two tensors you created in 3 using matrix multiplication
result = tf.matmul(random_tensor1, tf.transpose(random_tensor2))
result
```

```python
# Or you can use the @ operator for matrix multiplication
result_alt = random_tensor1 @ tf.transpose(random_tensor2)
result_alt
```

```python
# Or you can use it like this
result_alt2 = tf.linalg.matmul(random_tensor1, random_tensor2, transpose_b=True)
result_alt2
```

```python
# Multiply the two tensors you created in 3 using dot product
dot_product = tf.reduce_sum(random_tensor1 * random_tensor2, axis=-1)
dot_product
```

```python
# Or you can use tf.tensordot for dot product
dot_product_alt = tf.tensordot(random_tensor1, random_tensor2, axes=[[1], [1]])
dot_product_alt
```

```python
# Create a tensor with random values between 0 and 1 with shape [224, 224, 3]

random_values_tensor = tf.random.uniform(shape=[224, 224, 3], minval=0, maxval=1)
random_values_tensor
```

```python
#  Find the min and max values of the tensor you created in 6 along the first axis

min_values = tf.reduce_min(random_values_tensor, axis=0)
max_values = tf.reduce_max(random_values_tensor, axis=0)

min_values, max_values
```

```python
# Created a tensor with random values of shape [1, 224, 224, 3] then squeeze it to change the shape to [224, 224, 3]
random_tensor_4d = tf.random.uniform(shape=[1, 224, 224, 3], minval=0, maxval=1)
squeezed_tensor = tf.squeeze(random_tensor_4d, axis=0)
squeezed_tensor.shape
```

```python
# Create a tensor with shape [10] using your own choice of values, then find the index which has the maximum value
tensor_10 = tf.constant([1, 3, 5, 7, 9, 2, 4, 6, 8, 0])
max_index = tf.argmax(tensor_10)
max_index
```

```python
# One-hot encode the tensor you created in 9
one_hot_encoded = tf.one_hot(tensor_10, depth=10)
one_hot_encoded
```
