## 1. Creating our own data set

We are going to use a simple yet powerful way of seeing if our deep learing model works.
The equation is simple `f(x) = 2x`.
![Equation](https://substackcdn.com/image/fetch/$s_!IeZt!,w_1200,h_675,c_fill,f_jpg,q_auto:good,fl_progressive:steep,g_auto/https%3A%2F%2Fsubstack-post-media.s3.amazonaws.com%2Fpublic%2Fimages%2Fc0070a50-7da3-43bb-903c-8249d7d08dcb_1600x608.jpeg)



```python
# First we going to create the dataset
X = [n for n in range(1000)]
y = [n * 2 for n in range(1000)]
```

```python
import tensorflow as tf

X = tf.constant(tf.expand_dims(X, axis=-1))
y = tf.constant(tf.expand_dims(y, axis=-1))

# Creating our model
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(1),

])

# Compile the model
model1.compile(loss=tf.losses.mae,
               optimizer=tf.optimizers.SGD(),
               metrics=["mae"])

# Fitting the model
model1.fit(X, y, epochs=200)
```

```python
# Predict the number from our model
input = tf.constant([1000])
input = tf.expand_dims(input, axis=-1)
model1.predict(input)
```
